```python
df = pd.read_csv('Bank_Customers.csv')
```

```python
df
```

```python
df.drop(['RowNumber','CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
X, y = df.drop(['Exited'], axis=1), df['Exited']
```

```python
le = preprocessing.LabelEncoder()
scaler = StandardScaler()
ohe = preprocessing.OneHotEncoder()
mms = MinMaxScaler()
```

```python
X['Balance'] = scaler.fit_transform(X['Balance'].values.reshape(-1,1))
X['Balance']
```

```python
X['Gender'] = le.fit_transform(X['Gender'])
X['Gender']
```

```python
X['CreditScore'] = mms.fit_transform(X[['CreditScore']])
X['CreditScore'].value_counts()
```

```python
X['Geography'].value_counts()
```

```python
X = pd.get_dummies(X, columns = ['Geography'])
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.30, random_state=0)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.utils import np_utils
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units=128, activation='relu'))
ann.add(Dense(units=128, activation='relu'))
ann.add(Dense(units=128, activation='relu'))
ann.add(Dense(units=128, activation='relu'))
ann.add(Dropout(0.2))

ann.add(Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=300, epochs=50)
```
